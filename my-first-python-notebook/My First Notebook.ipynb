```python
import pandas as pd
import altair as alt
```

```python
committee_list = pd.read_csv('https://first-python-notebook.readthedocs.io/_static/committees.csv')

contribs = pd.read_csv('https://first-python-notebook.readthedocs.io/_static/contributions.csv')
```

```python
contribs.info()
```

```python
committee_list.info()
```

```python
committee_list['prop_name'].value_counts()
```

```python
my_prop = 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.'
```

```python
my_committees = committee_list[committee_list.prop_name == my_prop]
```

```python
my_committees.committee_position.value_counts()
```

```python
df = pd.merge(my_committees,contribs,on='calaccess_committee_id')

df.head()
```

```python
df.amount.sum()
```

```python
support = df[df.committee_position == 'SUPPORT']
support
```

```python
support.amount.sum() / df.amount.sum()
```

```python
oppose = df[df.committee_position == 'OPPOSE']
```

```python
oppose.amount.sum()
```

```python
support.sort_values('amount',ascending=False)
```

```python
df.groupby('contributor_state').amount.sum().reset_index().sort_values('amount',ascending=False)
```

```python
#PANDAS GROUPBYS do not include values where either column has nulls
#Use fillna to put in empty string to avoid this

df.fillna('').groupby(['contributor_firstname','contributor_lastname']).amount.sum().reset_index().sort_values('amount',ascending=False)
```

```python
top_donors = df.fillna('').groupby(['contributor_firstname','contributor_lastname','committee_position']).amount.sum().reset_index().sort_values('amount',ascending=False)
```

```python
top_donors.head()
```

```python
alt.Chart(top_donors.head()).mark_bar().encode(
    x='amount',
    y='contributor_lastname'
)
```

```python
alt.Chart(top_donors.head(10)).mark_bar().encode(
    x='amount',
    y=alt.Y('contributor_lastname', title='Name', sort='-x'),
    color='committee_position'
)
```
